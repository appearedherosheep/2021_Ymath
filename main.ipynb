```python
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
 
```

```python
# 임의의 데이터 만들기
def make_samples(n=1000):
    a = np.random.uniform(1, 8, (n, 1))
    b = np.random.uniform(1, 8, (n, 1))
    
    y = 5 * a + 2 * b + 40
    X = np.append(a, b, axis = 1)
    
    X = X / np.amax(X, axis = 0)
    y = y / 100
    
    return X, y
```

```python
X, y = make_samples()
X[:10]
```

```python
# 신경망 구현 
model = Sequential()
model.add(Dense(3, input_dim=2, activation='sigmoid')) 
model.add(Dense(1, activation='sigmoid'))
# Dense(출력 노드 수, input_dim = 입력 노드 수, activation=활성화함수)
```

```python
model.compile(optimizer='sgd', loss='mse')  # 회귀
# compile(optimizer=경사하강법, loss=손실함수)

# model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) 분류 
```

```python
# 학습
results = model.fit(X,y, epochs=50)
```

```python
# 오차 시각화 
plt.plot(results.history['loss'])
```
